```python
#Visualizations
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline

#Global API
x= np.arange(-10, 11)
plt.figure(figsize=(12,6))
plt.title('My first plot!')
plt.plot(x,x**2)  #plotting x-square graph
plt.plot(x,-1*(x**2))
```

```python
plt.figure(figsize=(12,6))  #creates a new figure with the given dimension
plt.title('nice plot')

#figure 1
plt.subplot(1,2,1) #rows, columns first plot
plt.plot(x,x**2)
plt.plot([0,0,0],[-10,0,100]) #this draws a line 
plt.legend(['X^2','Vertical Line']) #this draws the scale
plt.xlabel('X') #setting the x label to X
plt.ylabel('X Squared')#setting the y label to X Squared

#figure 2
plt.subplot(1,2,2) #rows, columns first plot
plt.plot(x,-1 * (x**2))
plt.plot([-10,0,10],[-50,-50,-50]) #this draws a line (x1,x2,x3),(y1,y2,y3)
plt.legend(['-X^2','Horizontal Line']) #this draws the scale
plt.xlabel('X') #setting the x label to X
plt.ylabel('X Squared')#setting the y label to X Squared
```
